#### Image classification of Cloths

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Flatten, Dense
```

```python
# Load Dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
# check sixe of dataset
len(x_train), len(y_train), len(x_test), len(x_train)
```

```python
# check dimension of dataset
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

##### Pre-process the data

```python
plt.imshow(x_train[1])
y_train[1]
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(x_train[i])
plt.show()
```

```python
# Mormalize dataset
x_train = x_train/255
x_test = x_test/255
```

```python
# build model
def build_model(model,x_train,y_train,x_test,y_test):
  model = model
  input_layer = Flatten(input_shape=(28,28))
  layer_1 = Dense(units=200,activation='relu')
  layer_2 = Dense(units=150, activation='sigmoid')
  output_layer = Dense(units=10, activation='softmax')
  model.add(input_layer), model.add(layer_1), model.add(layer_2), model.add(output_layer)
  model.compile(optimizer='SGD', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
  result = model.fit(x_train, y_train, epochs=20, validation_data=(x_test,y_test), verbose=1)
  return (result,model)
```

```python
# train model
history, model = build_model(Sequential(),x_train,y_train,x_test,y_test)
```

```python
# Evaluate model
df = pd.DataFrame(history.history)
df
```

```python
# Predict with model
y_pred = model.predict(x_test)
```

```python
y_pred[5]
```

```python
# Evaluating individual matrixes
np.argmax(y_pred[5]), y_test[5]
```

```python
# Display some images
fig, axes = plt.subplots(ncols=10, sharex=False,
    sharey=True, figsize=(20, 10))
for i in range(10):
    axes[i].set_title(str(y_test[i])+"-"+str(np.argmax(y_pred[i])))
    axes[i].imshow(x_test[i])
    axes[i].get_xaxis().set_visible(False)
    axes[i].get_yaxis().set_visible(False)
plt.show()
```

```python
# Evaluation matrix
confusion_matrix = tf.math.confusion_matrix(labels=y_test,predictions=np.argmax(y_pred, axis=1))
confusion_matrix
```

```python
sns.set(rc={'figure.figsize':(15,7)})
sns.heatmap(confusion_matrix,annot=True,cmap='PuBu_r',fmt='g')
plt.xlabel("Predicted digit")
plt.ylabel("Actual  digit")
```
